# Bar Chart

- A Bar chart / graph represent chategorical data with rantangular bars. The height of the bar corresponds to the value that they represent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

# Data Generation

```python
time_data = {'facebook': 120, 'youtube': 90, 'instagram':40, 'snapchat':40}
social_media = list(time_data.keys())
time_spend = list(time_data.values())
```

## Visualization

```python
plt.bar(social_media,time_spend)
```

```python
# Excercise 1 
plt.figure(figsize=(12, 8))   # Lengh & Height
plt.bar(social_media,time_spend, color= 'blue', width=0.5)
plt.xlabel('Social Media')
plt.ylabel ('Time in Minutes')
plt.title ('Average Time Spend in Social Media')
```

```python
# Excercise 2
style.use('ggplot')  # as shadow
plt.figure(figsize=(12, 8))   # Lengh & Height
plt.bar(social_media,time_spend, color= 'orange', width=0.5)
plt.xlabel('Social Media')
plt.ylabel ('Time in Minutes')
plt.title ('Average Time Spend in Social Media')
```

```python
# Excercise 3
style.use('ggplot')  # as shadow
plt.figure(figsize=(12, 8))   
color = ['royalblue','green','purple','red']  # diffent color for diffent graph
plt.bar(social_media,time_spend, color= color, width=0.5)
plt.xlabel('Social Media')
plt.ylabel ('Time in Minutes')
plt.title ('Average Time Spend in Social Media')
```

# Multiple BarChart

### Data For Visualization

```python
fb = (120, 130, 128, 100, 70)  # Appox. Minutes spent in facebook
yt = (80, 50, 70, 80, 65)
ig = (50, 40, 45, 35,50)
sc = (20, 50, 30, 15, 25)
```

```python
xpos = np.arange(len(fb))  # x position
xpos
```

```python
# Excercise 1
style.use('ggplot')
plt.figure(figsize=(8,6))
barWidth = 0.2
plt.bar(xpos, fb, color='royalblue', width=barWidth, label='FB')
```

```python
# Excercise 2
style.use('ggplot')
plt.figure(figsize=(8,6))
barWidth = 0.2
plt.bar(xpos, fb, color='royalblue', width=barWidth, label='FB')
plt.bar(xpos, yt, color='red', width=barWidth, label='YT')      # Data Overlabpped in same graph, the position can be changed by adding xpos+0.2 etc.
```

```python
# Excercise 3
style.use('ggplot')
plt.figure(figsize=(8,6))
barWidth = 0.2
plt.bar(xpos, fb, color='royalblue', width=barWidth, label='FB')
plt.bar(xpos+0.2, yt, color='red', width=barWidth, label='YT')     # Reposition 0.2 from fb
plt.bar(xpos+0.4, ig, color='purple', width=barWidth, label='IG')  # Reposition 0.4 from fb
plt.bar(xpos+0.6, sc, color='orange', width=barWidth, label='SC')  # Reposition 0.6 from fb
plt.xlabel('Day')
plt.ylabel('Time Spend in Minutes')
plt.legend()
```

- xticks() Function. The annotate() function in pyplot module of matplotlib library is used to get and set the current tick locations and labels of the x-axis.

```python
# Excercise 4 : Add Day in the graph
style.use('ggplot')
plt.figure(figsize=(8,6))
barWidth = 0.2
plt.bar(xpos, fb, color='royalblue', width=barWidth, label='FB')
plt.bar(xpos+0.2, yt, color='red', width=barWidth, label='YT')     # Reposition 0.2 from fb
plt.bar(xpos+0.4, ig, color='purple', width=barWidth, label='IG')  # Reposition 0.4 from fb
plt.bar(xpos+0.6, sc, color='orange', width=barWidth, label='SC')  # Reposition 0.6 from fb
plt.xticks(xpos+0.3,('Monday','Tuesday','Wednesday','Thursday', 'Friday'))  # xtricks applied for x axis
plt.xlabel('Day')
plt.ylabel('Time Spend in Minutes')
plt.title('Time Spend in Social Media')
plt.legend()
```

```python
fb = (120, 130, 128, 100, 70)  
yt = (80, 50, 70, 80, 65)
ig = (50, 40, 45, 35,50)
x = ['Monday','Tuesday','Wednesday', 'Thursday','Friday']
xpos = np.arange(len(x))
barWidth= 0.4
plt.figure(figsize=(6,4))
plt.bar(xpos, fb, color='royalblue', width=barWidth, label='FB')
plt.bar(xpos, yt, color='orange', width=barWidth, label='YT')
plt.bar(xpos, ig, color='purple', width=barWidth, label='IG')
plt.xlabel('Day')
plt.ylabel('Time Spend in Minutes')
plt.title('Time Spend in Social Media')
plt.xticks(xpos, x)
plt.legend()
```

```python
fb = (120, 130, 128, 100, 70)  
yt = (80, 50, 70, 80, 65)
ig = (50, 40, 45, 35,50)
x = ['Monday','Tuesday','Wednesday', 'Thursday','Friday']
xpos = np.arange(len(x))
barWidth= 0.4
plt.figure(figsize=(6,4))
plt.bar(xpos, fb, color='royalblue', width=barWidth, label='FB')
plt.bar(xpos, yt, bottom=fb, color='red', width=barWidth, label='YT')  # bottom mean the 2nd data will be on first data
plt.bar(xpos, ig, bottom=yt, color='purple', width=barWidth, label='IG') 
plt.xlabel('Day')
plt.ylabel('Time Spend in Minutes')
plt.title('Time Spend in Social Media')
plt.xticks(xpos, x)
plt.legend()
```

# Plot from CSV File

```python
df = pd.read_csv('1002_salesData.csv')
df
```

```python
team = df['Team']
sales = df['Actual Sales']/10000000
```

### Vertical Barchart

```python
fig = plt.figure(figsize=(10, 8))
plt.bar(team, sales)    
plt.xlabel('Sales Team')
plt.ylabel('Sales BDT in Cr')
plt.title('Sales Analysis')
plt.show()
```

### Horizontal Barchart

```python
fig = plt.figure(figsize=(10, 8))
plt.barh(team, sales, color='blue')              # barh : Horizontal Barchart
plt.xlabel('Sales BDT in Cr')
plt.ylabel('Sales Team')
plt.title('Sales Analysis')
plt.show()
```
