```python
df = sns.load_dataset('car_crashes')
df.head()
```

# Distribution Plots

## Distribution Plot

```python
sns.distplot(df['not_distracted'])
```

```python
sns.distplot(df['not_distracted'], kde=False)  # KDE: Kernel Density Estimation, which is false
```

```python
sns.distplot(df['not_distracted'], kde=False, bins=25)
```

## Joint Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=df)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg') # adding regression line 
```

```python
sns.jointplot(x='speeding', y='alcohol', data=df, kind='kde') # adding kde
```

```python
sns.jointplot(x='speeding', y='alcohol', data=df, kind='hex') # adding hexagoan distribution
```

## KDE Plot

```python
sns.kdeplot(df['alcohol'])
```

# Styling

```python
sns.set_style('white')  # style white / darkgrid

plt.figure(figsize=(10,6))

sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')    
```

```python
sns.set_style('darkgrid')  # style white / darkgrid

plt.figure(figsize=(10,6))

sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
plt.show()
```

```python
sns.set_style('whitegrid')  # Style

plt.figure(figsize=(10,6))

sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
plt.show()
```

```python
sns.set_style('dark')  # style 

plt.figure(figsize=(10,6))

sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
plt.show()
```

```python
sns.set_style('ticks')  # style
plt.figure(figsize=(8,6))
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
plt.show()
```

```python
sns.set_style('ticks')
plt.figure(figsize=(10,8))
sns.set_context('paper', font_scale=1.6)  # axis text customization: available font: paper, talk & poster
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')

plt.show()
```

```python
sns.set_style('ticks')
plt.figure(figsize=(10,8))
sns.set_context('talk', font_scale=1.4)  
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
sns.set_context('poster', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
plt.show()
```

```python
sns.set_style('white')
plt.figure(figsize=(10,8))
sns.set_context('paper', font_scale= 1.4)
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
sns.despine(left=True)  # Remove/ Despine left axis line
```

```python
sns.set_style('white')
plt.figure(figsize=(10,8))
sns.set_context('paper', font_scale= 1.4)
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
sns.despine(left=False)  # Showing left axis line
```

```python
sns.set_style('white')
plt.figure(figsize=(10,8))
sns.set_context('paper', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
sns.despine(left=True, bottom=True)  # bollom axis line off
```

# Categorical Plots

## Bar Plot

```python
 sns.barplot(x='sex', y='total_bill', data=df2)
```

```python
sns.barplot(x='sex', y='total_bill', data=df2, estimator=np.median) # base of median
```

```python
sns.barplot(x='sex', y='total_bill', data=df2, estimator=np.std) # Base of std
```

```python
sns.barplot(x='sex', y='total_bill', data=df2, estimator=np.var) # Base of variance
```

```python
sns.barplot(x='sex', y='total_bill', data=df2, estimator=np.cov) # Base of covariance
```

# Count Plot

```python
sns.countplot(x='sex', data=df2)
```

# Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=df2, hue='sex')
plt.legend(loc=0)
```

# Violin Plot

- Violin plot is the combination of box & kde plot

```python
sns.violinplot(x='day', y='total_bill', data=df2, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=df2, hue='sex', split=True)
```
