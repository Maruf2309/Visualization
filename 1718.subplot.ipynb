```python
# Finding Feature List - Only Category
feature = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod']                                                # remove "tenure" becuse its continuous variable, Keep only categorical variable

# Write a fun that is enumerate
```

### Enumerate

- It will return tuble with index number & relative col name
- Find tolal 16 columns from 0 to 15
- It will be helpful for doing the iteration

```python
enumerate(feature)
```

```python
list(enumerate(feature))
```

### 6 rows & 3 columns

```python
for i in enumerate(feature):
    plt.subplot(6,3,i[0]+1)
    sns.countplot(x=i[1], hue='Churn', data=df)             # x=i[1], i= 0 & values like features, parter , contact ect
```

```python
plt.figure(figsize=(15, 25))
for i in enumerate(feature):
    plt.subplot(6,3,i[0]+1)
    sns.countplot(x=i[1], hue='Churn', data=df)  
```

### 4 Rows & 4 Columns

```python
plt.figure(figsize=(15, 15))
for i in enumerate(feature):
    plt.subplot(4,4,i[0]+1)
    sns.countplot(x=i[1], hue='Churn', data=df) 
    plt.xticks(rotation=45)    # for removeming overlapping text ratation 45 degree
```
